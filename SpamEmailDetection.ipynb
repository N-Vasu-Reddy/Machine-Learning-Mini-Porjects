```python
import numpy as np
import pandas as pd
```

```python
#importing LogisticRegression model
from sklearn.linear_model import LogisticRegression
```

```python
#importing accuracy score function
from sklearn.metrics import accuracy_score
```

```python
#Import tain_test_split function
from sklearn.model_selection import train_test_split
```

```python
#To convert all strings into numeric we import TfidfVectorizer from feature_extraction.text in sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#importing the dataset
raw_mail_data = pd.read_csv('/content/drive/MyDrive/New folder/spam.csv',encoding='latin1')
raw_mail_data
```

```python
raw_mail_data= raw_mail_data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
raw_mail_data.rename(columns={'v1' : 'Category','v2':'Message'}, inplace =True)
```

```python
#replacing all the null values with null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the first 5 rows
mail_data.head()
```

```python
#checking total noof rows and columns of the dataset
mail_data.shape
```

**Label Encoding**

```python
#Lael spam emails as 0 and ham emails as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

```python
#seperating the data as text and label
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train , x_test, y_train ,y_test =train_test_split(x ,y ,test_size = 0.2, random_state= 2)
print(x.shape, x_train.shape , x_test.shape)
```

```python
#Transforming the text data to feature vectors that can be used as input to logistic regression
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english')
```

```python
x_train_features =feature_extraction.fit_transform(x_train)
```

```python
x_test_features = feature_extraction.transform(x_test)
#in test_data we do not fit the data.
'''
*The reason we don't fit the model on the test data is to ensure an unbiased evaluation of its performance.
*The test data serves as a kind of "unseen" data that the model hasn't encountered during training.
*If we used the test data to train the model as well, it would essentially already know the answers,
and we wouldn't get a true measure of how well it generalizes to new, unseen data.

**Hence we do not fit the data in testing data
'''
```

```python
x_test.head(20)
```

```python
y_train
```

```python
#Here the dtype of y_train is object(which can contain NUMERICS)
#so we have to convert t_train and y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

**Training the model**

```python
model = LogisticRegression()
```

```python
#Training the logistic regression model with the training data
model.fit(x_train_features,y_train)#model.fit(training input, training output)
```

```python
#Evaluating the trained model
prediction_on_training_data = model.predict(x_train_features)
#Finding the accuracy score on training data
accuracy_on_training_data = accuracy_score(y_train,prediction_on_training_data)#accuracy_score(training output,predicted training input )
print("The Accuracy score on training data is ",accuracy_on_training_data)
```

```python
#Evaluating the testing model
prediction_on_testing_data = model.predict(x_test_features)
#Finding the accuracy score on testing data
accuracy_on_testing_data = accuracy_score(y_test,prediction_on_testing_data)
print("The Accuracy score on testing data is ",accuracy_on_testing_data)
```

```python
#building a predictive system
input_mail = ['Welcome to Rummy Circle. You have won a login of 2500']
#Converting the feature vectors (Numerical format)
input_mail_num = feature_extraction.transform(input_mail)
```

```python
#Making prediction
prediction = model.predict(input_mail_num)
if(prediction[0] ==0 ):
  print('Spam Email')
else:
  print('Ham Email')
```
